```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/prevalence-of-depression-males-vs-females.csv')
```

```python
df.drop(columns=['Continent','Population (historical estimates)'])
```

```python
df.rename(columns={'Entity': 'Country',
                 'Population (historical estimates)': 'Population',
                 df.columns[3] : 'Male',
                 df.columns[4] : 'Female'}, inplace=True)
df.drop(columns='Code', inplace=True)
#transformando variables!
```

```python
df.columns
```

```python
df.info()
```

```python
df = df[(df['Male'].notnull())&(df['Female'].notnull())]
```

```python
df = df.drop(columns='Continent')
df = df.drop(columns='Population')
```

```python
df.groupby('Country').describe()['Male']['mean']
```

```python
df.groupby('Country').describe()['Female']['mean']
```

```python
df[df['Country'].str.contains('WHO')]['Country'].unique()
dfcont = df[df['Country'].str.contains('WHO')].copy()
dfcont.rename(columns={'Country': 'Continent'}, inplace=True)
```

```python
df[df['Country'].str.contains('Bank')]['Country'].unique()
dfincome = df[df['Country'].str.contains('Bank')].copy()
dfincome.rename(columns={'Country': 'Income'}, inplace=True)
```

```python
df = df[~df.Country.str.contains("WHO")]
df = df[~df.Country.str.contains("Bank")]
```
