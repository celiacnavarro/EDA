```python
import numpy as np
import pandas as pd
```

```python
prevalencias = pd.read_csv('data/prevalence-by-mental-and-substance-use-disorder.csv')
```

```python
prevalencias
```

```python
disorders = ['Schizofrenia', 'Bipolar disorder', 'Eating disorders', 'Anxiety disorders', 'Drug use disorders', 'Depressive disorders', 'Alcohol use disorders']
for col in prevalencias.columns:
    for i in disorders:
        if i in col:
            prevalencias.rename(columns={col:i}, inplace=True)

```

```python
prevalencias.rename(columns={'Entity': 'Country'}, inplace=True)
prevalencias.drop(columns='Code', inplace=True)
#transformando variables!
```

```python
prevalencias.columns
```

```python
for i in range(2,9):
    new_name = prevalencias.columns[i].split()[2]
    prevalencias.rename(columns={prevalencias.columns[i]: new_name}, inplace=True)
    print(new_name)

#cambiando los nombres de las variables
```
